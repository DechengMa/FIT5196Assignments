## Task 1 - Sparse Feature Generation

We use the tabula libraries wrapper function to convert the existing pdf into csv format. 

```python
w.convert_into('Group017.pdf', 'urls.csv', output_format="csv", pages='all') #converts pdf to csv
```

Next, we create a dataframe "urls" which consists of two columns, the document name and its url. We then convert this to a dictionary, witht he filename as key and url as the value.

```python
urls=pd.read_csv('urls.csv', delimiter=',') #reading the above csv and storing it in a dataframe.
url_dict = urls.set_index('filename')['url'].to_dict() #creating a dictionary with key= filename and value= url
del url_dict['filename'] #deleting the first element of the dictionary(the heading of the dataframe)
print(url_dict)
```

Next, we access each pdf using the url provided and download the pdf files. 

```python
for key, value in url_dict.items():
    names=Path(key) #specifying the path 
    link= value #specifying the url
    reponse= requests.get(link) #Using the requests library to access the link
    names.write_bytes(reponse.content) #downloading the pdf files 
```

We then convert each pdf file into txt using the pdfminer library 

```python
for key, value in url_dict.items():
    temp=re.findall(r'PP\d+', key)  #getting the file name without the extension
    for item in temp:
        !pdf2txt.py -o "$item".txt "$item".pdf #converting pdf to txt
```

We then created a list of all the file names with the txt extension

```python
doc_names=[]
for key in url_dict.keys():
    temp=re.findall(r'PP\d+', key)
    for item in temp:
        doc_names.append(item+".txt")  
```

#### Normalization

We next normalized each piece of text corresponding to the txt files- this means that we converted words to lowercase except the capital words appearing in the middle of a sentence/line.

```python
normalized_files=dict()
regex = r'1 Paper Body([\s\S]*?)2 References'  #specifying regex to extract relevent portions of text (from Paper Bodies till References not including these headings)
for item in doc_names:
    text= open(item, 'r') #opening each txt file to read
    text= text.read()
    match = re.findall(regex, text) #extracting relevent portions of text
    sample = match[0] #only one such portion exists in each txt file. Thus length is of sample is never greater than 1
    sample=re.sub('\?', '', sample)
    normalizedtext = ""
    sent_tokenize_list = sent_tokenize(sample) #tokenizing sentences
    normalizedsentencelist = []
    for sentence in sent_tokenize_list:
        normalizedsentance = sentence.replace(sentence.split(' ', 1)[0],sentence.split(' ', 1)[0].lower()) #converting to lower case if the first character of the sentence is capital
        normalizedsentencelist.append(normalizedsentance)
        normalizedtext+= normalizedsentance  #putting all the converted sentences together
    normalized_files[item]=(normalizedtext) #dictionary where the key is the file name and the value is the normalized text
```

#### Tokenization

We used the provided regular expression and the RegexpTokenizer to tokenize the words in each doument. We have a bag of words for each doument stored in the "tokens_dict".

```python
tokens_dict=dict()
tokenizer = RegexpTokenizer(r"[A-Za-z]\w+(?:[-'?]\w+)?")
for key, item in normalized_files.items():
    tokens = tokenizer.tokenize(item) #tokenizing each file
    tokens_dict[key]=tokens #creating a dictionary with key as the file name and values as a list of tokens of that file
```

```python
words = list(chain.from_iterable(tokens_dict.values())) #finding the total number of tokens 
vocab = set(words) #finding the number of unique words
print ("Vocabulary size: ",len(vocab),"\nTotal number of tokens: ", len(words))
```

```python
print(vocab)
```

#### Removing Stop words
##### Context Independent 
First we remove the stopwords specified in the stopwords file. 

```python
stopwords = []
with open('stopwords_en.txt', 'r') as f: #opening the provided stop words file
    stopwords = f.read().splitlines() #reading the stop words from the file
for name in tokens_dict.keys():
    tokens_dict[name] = [w for w in tokens_dict[name] if w not in stopwords] #
```

```python
words = list(chain.from_iterable(tokens_dict.values())) #finding the total number of tokens 
vocab = set(words) #finding the number of unique words
print ("Vocabulary size: ",len(vocab),"\nTotal number of tokens: ", len(words))
```

##### Context Dependent 

We want to remove context dependent stop words with a threshold of 95%. There are 200 documents- we need to remove the words that appear in 95% of the 200 documents. Therefore if the document count of a word is greater then 190, we remove it.

```python
words= list(chain.from_iterable([set(value) for value in tokens_dict.values()])) #finds unique words
fd= FreqDist(words) #finds the number of documents the word appears in 
for name in tokens_dict.keys():
    tokens_dict[name] = [w for w in tokens_dict[name] if not fd[w]>190] #if the word occurs in more than 190 documents it becomes a stop word
        
```

```python
words = list(chain.from_iterable(tokens_dict.values())) #finding the total number of tokens 
vocab = set(words) #finding the number of unique words
print ("Vocabulary size: ",len(vocab),"\nTotal number of tokens: ", len(words))
```

##### Rare Tokens

Similarly, we want to remove rare tokens with a threshold of 3%. There are 200 documents- we need to remove the words that appear in 3% or less of the 200 documents. Therefore if the document count of a word is less than 6, we remove it.

```python
words=list(chain.from_iterable([set(value) for value in tokens_dict.values()])) #finds unique words
fd= FreqDist(words) #finds the number of documents the word appears in 
for name in tokens_dict.keys():
    tokens_dict[name] = [w for w in tokens_dict[name] if not fd[w]<6] #if the word occurs in less than 6 documents it becomes a stop word
```

```python
words = list(chain.from_iterable(tokens_dict.values())) #finding the total number of tokens 
vocab = set(words) #finding the number of unique words
print ("Vocabulary size: ",len(vocab),"\nTotal number of tokens: ", len(words))
```

##### Length less than 3

We also want to remove words that have length less than 3. The following code can be used to do this.

```python
#removing words of length less than 3
for name in tokens_dict.keys():
    tokens_dict[name] = [w for w in tokens_dict[name] if len(w)>=3] #if the word length is less than 3 it is not included
```

```python
words = list(chain.from_iterable(tokens_dict.values())) #finding the total number of tokens 
vocab = set(words) #finding the number of unique words
print ("Vocabulary size: ",len(vocab),"\nTotal number of tokens: ", len(words))
```

#### Bigrams

Next, we want to find the 200 most common Bigrams. Bigrams in this context are a pair of consecutive written words. We use the following code to find the 200 Bigrams. 

```python
fd = FreqDist(words) #finds the frequency of each word
bigrams = ngrams(words, n = 2) #finds the ngrams of length 2
fdbigram = FreqDist(bigrams) #finds the frequency of each bigram found above
bigrams_200=fdbigram.most_common(200) #finds the top 200 bigrams 
bigrams_final=[]
for item in bigrams_200:
    bigrams_final.append("__".join(item[0])) #seperates the two words by double underscore
print(bigrams_final)
```

#### Stemming

Stemming is the process of identification and removal of prefixes, suffixes, and pluralisation, and leaves you with a stem. The following code uses the PorterStemmer to carry out this process. 

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
for name in tokens_dict.keys():
    list1=[]
    for w in tokens_dict[name]:
        if w.isupper(): #Checking to see if the entire word is capital
            w=stemmer.stem(w) #stemming the word
            w=w.upper() #converts the entire word back to capital
        elif w[0].isupper(): #Checking to see if the first letter of the word is capital
            w=stemmer.stem(w) #stemming the word
            w=w.capitalize() #converts the first letter of the word back to capital
        else:
            w=stemmer.stem(w) #stemming the word
        list1.append(w) #creating a list of the stemmed words on this particular document 
    tokens_dict[name]=list1 #changing the dictionary so that the key is the document name and the value is the list of stemmed words
    
```

```python
words = list(chain.from_iterable(tokens_dict.values())) #finding the total number of tokens 
vocab = set(words) #finding the number of unique words
print ("Vocabulary size: ",len(vocab),"\nTotal number of tokens: ", len(words))
```

#### Final Vocab List

This is a list of all the unigram tokens along with the list of 200 bigrams that we found above in alphabetical order

```python
final_vocab= sorted(list(vocab)+bigrams_final) #total vocab containing the bigrams as well
```

#### Vocabulary index file

This file contains the bigrams and unigrams tokens in the following format: "token_string:token_index".

```python
i=0 #initializing the index value
vocab_dict=dict()
out_file=open('Group017_vocab.txt', 'w') #opening a text file to write in
for item in final_vocab: 
    out_file.write(item+":"+str(i)+"\n") #writing to the output file in the required format
    vocab_dict[item]=i #creating a dictionary of the words and their index
    i+=1
```

#### Sparse count vectors file

Each line in this txt file contains the frequency of the tokens and the tokens index of one of the papers in the following format:
"paper_id, token1_index:token1_wordcount, token2_index:token2_wordcount, etc."


```python
out_file2=open('Group017_count_vectors.txt', 'w')
for name in tokens_dict.keys():
    temp=re.findall(r'PP\d+', name)
    for item in temp:
        doc_name=item  
    fd=FreqDist(tokens_dict[name]) #finding the frequency of each word
    out_file2.write(doc_name+",")
    for k, v in fd.items():
        out_file2.write(str(vocab_dict[k])+":"+str(v)+",") #mapping each word to its index in the "vocab_dict" dictionary
    out_file2.write('\n')
```

## Task 2 - Statistics Generation

In order to go about generating the required Statistics, we started by removing the title, abstract and authors using regular expressions

```python
titleregex = re.compile(r'([\s\S]*)Authored') #regular expression for title
authorregex = 'by:([\s\S]*?)Abstract' #regular expression for author
abstractregex = 'Abstract([\s\S]*?)1 Paper Body' #regular expression for abstract

stopwords_list=[] 
stopwords=open('stopwords_en.txt', 'r') #reading stopwords
stopwords=stopwords.readlines()
for item in stopwords:
    stopwords_list.append(re.sub('\n','',item))
stopwords_set = set(stopwords_list) # removing duplicates from stopwords



# stopwords_list
list_of_authors = []
list_of_titles = []
list_of_abstracts = []
title_abstract = dict()
for item in doc_names:
    text= open(item, 'r', encoding= 'utf-8') #reading each text file
    text= text.read()
    title = re.search(titleregex, text).group(1) # finding title
    list_of_titles.append(title)# adding title to list
    authors = re.search(authorregex, text).group(1) # finding authors
    authors = authors.strip().splitlines()
    for author in authors: 
        if not(author==''):
            list_of_authors.append(author) # adding each author to list
    abstract = re.search(abstractregex, text).group(1).strip() # finding abstract
    list_of_abstracts.append(abstract) #adding abstract to list
    title_abstract[title]=abstract # creating dictionary with title and abstract
#     print(abstract)
# print(list_of_abstracts)
```

We then performed Tokenization of the title, abstract and authors, followed by the removal of stop words

```python
tokenizer = RegexpTokenizer(r"[A-Za-z]\w+(?:[-'?]\w+)?") # using given regex to tokenize
normalizedabstractlist = []
tokenized_abstract_list=[]

for abstract in list_of_abstracts:
    abstract = re.sub('\n', '', abstract) # removing all \n
    abstract = abstract.strip() # removing all trailing white spaces
    abstract_tokenize_list = sent_tokenize(abstract) # sentence segmentation
    tok = ''
    for token in abstract_tokenize_list:
        normalizedtoken = token.replace(token.split(' ', 1)[0],token.split(' ', 1)[0].lower()) # normalizing only the starting word of a sentence
#         print(normalizedtoken)
        tok+=normalizedtoken
        tokenized_abstract_list+=tokenizer.tokenize(normalizedtoken)
    normalizedabstractlist.append(tok) # adding normalized tokens to a list
# print(tokenized_abstract_list)


stopped_tokens_abstract = [w for w in tokenized_abstract_list if w not in stopwords_set] # reomving stopwords from list of tokens
# print(stopped_tokens_abstract)
```

```python
list_of_titles_normalized=[]
title_token_list = []

for title in list_of_titles: 
    title=re.sub('\n','',title)
    title=title.strip() # removing all traliling white spaces
    list_of_titles_normalized.append(title.lower()) 
    title_tokens = tokenizer.tokenize(title.lower())
    title_token_list+=(title_tokens)
    stopped_tokens_title = [w for w in title_token_list if w not in stopwords_set] # removing all stopwords from tokenized list

stopped_tokens_title
```

We next wrote code to find the most common words among the title, abstract and authors.

```python
#creating most common title words list
counts_titles = collections.Counter(stopped_tokens_title)
most_common_title_words = []
for x in counts_titles.most_common(10):
    most_common_title_words.append(x[0])
```

```python
#creating most common abstract words list
counts_abstract = collections.Counter(stopped_tokens_abstract)
counts_abstract.most_common()
most_common_abstract_words = []
for x in counts_abstract.most_common(10):
    most_common_abstract_words.append(x[0])
```

```python
#creating most common authors
counts_author = collections.Counter(list_of_authors)
counts_author.most_common()
most_common_author_words = []
for x in counts_author.most_common(10):
    most_common_author_words.append(x[0])
```

Finally, we created a dataframe containing the required information and saved this dataframe as a CSV file.

```python
df = pd.DataFrame() # new dataframe
# populating dataframe
df['top10_terms_in_abstracts'] = most_common_abstract_words
df['top10_terms_in_titles'] = most_common_title_words
df['top10_authors'] = most_common_author_words
# df.index = df['top10_terms_in_abstracts']
df.to_csv('Group017_stats.csv', index=False) # writing to csv
```
