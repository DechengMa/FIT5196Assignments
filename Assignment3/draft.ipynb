### distance_to_customer_KM

Here we need to use the djkistra algorithm

```python
branches= pd.read_csv("branches.csv")
```

```python
nodes=pd.read_csv("nodes.csv")
```

```python
edges= pd.read_csv("edges.csv")
```

```python
branches.head()
```

```python
nodes.head()
```

```python
edges.head()
```

```python
branch_nodes=pd.merge(branches, nodes, how='left', left_on=['branch_lat','branch_lon'], right_on = ['lat','lon'])
branch_nodes
```

```python
branch_nodes=branch_nodes.drop(['lat', 'lon'], axis=1)
```

```python
branch_nodes.rename(columns={'node':'branch_node'}, inplace=True)
```

```python
branch_nodes
```

```python
missing=pd.merge(missing, nodes, how='left', left_on=['customer_lat','customer_lon'], right_on = ['lat','lon'])
missing=missing.drop(['lat', 'lon'], axis=1)
missing.rename(columns={'node':'customer_node'}, inplace=True)
missing
```

```python
missing.head()
```

```python
missing=pd.merge(missing, branch_nodes, how='left', left_on=['branch_code'], right_on = ['branch_code'])
```

```python
missing.head()
```

```python
nodes_list=list(nodes['node'])
```

```python
edges_list=list(zip(edges.u, edges.v, edges['distance(m)']))
```

```python
import networkx as nx

G=nx.Graph()
G.add_nodes_from(nodes_list)
G.add_weighted_edges_from(edges_list)
```

```python
source_list=list(missing.loc[(missing['distance_to_customer_KM'].isnull())]['branch_node'])
target_list=list(missing.loc[(missing['distance_to_customer_KM'].isnull())]['customer_node'])
```

```python
dist=list()
for i in range(len(source_list)):
    dist.append((nx.dijkstra_path_length(G, source_list[i], target_list[i]))/1000)
len(dist)
```

```python
dist[0]
```

```python
missing['distance_to_customer_KM'][missing['distance_to_customer_KM'].isnull()]=dist
```

```python
missing['distance_to_customer_KM'].isnull().sum()
```

Thus, missing values of distance have been filled.

```python
import datetime
import calendar
```

```python
day=[]
for item in date:
    if datetime.datetime.strptime(item, '%Y-%m-%d').strftime('%A') in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']:
        day.append(0)
    else:
        day.append(1)
```

```python
datetime.datetime.strptime(time[0], '%H:%M:%S').time()
```

```python
if datetime.datetime.strptime(time[0], '%H:%M:%S').time()> datetime.datetime.strptime("08:00:00", '%H:%M:%S').time() and datetime.datetime.strptime(time[0], '%H:%M:%S').time()<=datetime.datetime.strptime("12:00:00", '%H:%M:%S').time():
    print(True)
else:
    print(False)
```

```python
def times(timestring):
    return datetime.datetime.strptime(timestring, '%H:%M:%S').time()
```

```python
tod=[]
for item in time:
    if times(item)>= times("08:00:00") and times(item)<=times("12:00:00"):
        tod.append(0)
    elif times(item)> times("12:00:00") and times(item)<=times("16:00:00"):
        tod.append(1)
    elif times(item)> times("16:00:00") and times(item)<=times("20:00:00"):
        tod.append(2)
    else:
        print("ERROR")
        break
```

```python
missing['TimeOfDay']=tod
```
